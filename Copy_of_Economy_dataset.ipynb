**Project name** : Predicting Prosperity: Economic Forecasting with KNN

**General description of data:**

CountryID: A unique identifier for each country.

Country: The name of the country.

Year: The year of the data record.

AMA Exchange Rate: The exchange rate based on the AMA system.

IMF-based Exchange Rate: The exchange rate used by the International Monetary Fund.

Population: The total population of the country for that year.

Currency: The currency used in the country.

Per Capita GNI: Gross National Income per capita (in USD).

Agriculture, Hunting, Forestry, Fishing (ISIC A-B): The value of output from agriculture, hunting, forestry, and fishing sectors.

Changes in Inventories: The change in the country's inventories (unsold goods, etc.).

Construction (ISIC F): The output from the construction sector.

Exports of Goods and Services: The total value of goods and services exported.

Final Consumption Expenditure: The total value of consumption expenditure by households and government.

General Government Final Consumption Expenditure: Government spending on goods and services.

Gross Capital Formation: Total investment in physical assets (including new buildings, machinery, etc.).

Gross Fixed Capital Formation: Investment in long-term physical assets, including acquisitions and disposals of valuables.

Household Consumption Expenditure: The total consumption expenditure by households, including spending on goods and services.

Imports of Goods and Services: The total value of goods and services imported.

Manufacturing (ISIC D): The output from the manufacturing sector.

Mining, Manufacturing, Utilities (ISIC C-E): The output from mining, manufacturing, and utility sectors.

Other Activities (ISIC J-P): Other economic activities (based on the International Standard Industrial Classification).

Total Value Added: The total value added by all sectors of the economy.

Transport, Storage, and Communication (ISIC I): The output from the transportation, storage, and communication sectors.

Wholesale, Retail Trade, Restaurants, and Hotels (ISIC G-H): The value of output from wholesale and retail trade, as well as restaurants and hotels.

Gross National Income (GNI) in USD: Total income generated by the country's residents, including income earned abroad.

Gross Domestic Product (GDP): The total market value of all final goods and services produced within the country.

```python
df.columns = df.columns.str.strip()

df = df.drop(columns=['Changes in inventories'])

```

Extra spaces were removed from column names to prevent errors when referencing them. The column "Changes in inventories" was dropped due to a large number of missing values and low relevance to the goal of predicting GDP growth

```python
numeric_cols = df.select_dtypes(include='number').columns

df[numeric_cols] = df.groupby('Country')[numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

Missing values in numeric columns were filled with the mean value for each country. This approach ensures that the imputation is contextually relevant to each country’s data, preventing bias from the global average.

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

For any remaining missing values that could not be filled by the country mean, the global mean was used as a fallback. This ensures that no data points are left as missing, maintaining a complete dataset for analysis.
